### 2.2 Code Ä‘áº§y Ä‘á»§

```lua
--[[
    WaveEmoteScript.lua
    Vá»‹ trÃ­: StarterPlayer > StarterPlayerScripts
    Chá»©c nÄƒng: PhÃ¡t Wave animation khi nháº¥n E
--]]

-- Services
local Players = game:GetService("Players")
local UserInputService = game:GetService("UserInputService")

-- Configuration
local WAVE_ANIMATION_ID = "rbxassetid://YOUR_ANIMATION_ID"  -- âš ï¸ THAY ID Cá»¦A Báº N
local TRIGGER_KEY = Enum.KeyCode.E

-- Variables
local player = Players.LocalPlayer
local waveTrack = nil
local isWaving = false

-- Function: Setup animation cho character
local function setupAnimation(character)
    local humanoid = character:WaitForChild("Humanoid")
    local animator = humanoid:WaitForChild("Animator")
    
    -- Táº¡o Animation object
    local waveAnim = Instance.new("Animation")
    waveAnim.AnimationId = WAVE_ANIMATION_ID
    waveAnim.Name = "WaveEmote"
    
    -- Load animation
    waveTrack = animator:LoadAnimation(waveAnim)
    waveTrack.Priority = Enum.AnimationPriority.Action
    waveTrack.Looped = false
    
    -- Event khi animation dá»«ng
    waveTrack.Stopped:Connect(function()
        isWaving = false
        print("Wave finished!")
    end)
    
    print("Wave animation loaded!")
end

-- Function: PhÃ¡t Wave animation
local function playWave()
    -- Kiá»ƒm tra Ä‘iá»u kiá»‡n
    if isWaving then
        print("Already waving!")
        return
    end
    
    if not waveTrack then
        warn("Wave track not loaded!")
        return
    end
    
    -- PhÃ¡t animation
    isWaving = true
    waveTrack:Play(0.1)  -- Fade in 0.1 giÃ¢y
    print("Waving!")
end

-- Setup khi character spawn
local function onCharacterAdded(character)
    setupAnimation(character)
end

-- Event: Character spawns
player.CharacterAdded:Connect(onCharacterAdded)

-- Setup cho character hiá»‡n táº¡i
if player.Character then
    onCharacterAdded(player.Character)
end

-- Event: Input handling
UserInputService.InputBegan:Connect(function(input, gameProcessed)
    -- Bá» qua náº¿u Ä‘ang chat hoáº·c trong GUI
    if gameProcessed then return end
    
    -- Kiá»ƒm tra phÃ­m E
    if input.KeyCode == TRIGGER_KEY then
        playWave()
    end
end)

-- Print hÆ°á»›ng dáº«n
print("================================")
print("   WAVE EMOTE LOADED!")
print("   Press E to wave!")
print("================================")
```

## BÆ°á»›c 2: Viáº¿t Script Idle System

### 2.1 Script Ä‘áº§y Ä‘á»§

```lua
--[[
    IdleVariationSystem.lua
    Vá»‹ trÃ­: StarterPlayer > StarterCharacterScripts
    Chá»©c nÄƒng: Random idle animations khi Ä‘á»©ng yÃªn
--]]

-- Configuration
local CONFIG = {
    -- Animation IDs (âš ï¸ THAY Báº°NG ID Cá»¦A Báº N)
    IdleBase = "rbxassetid://YOUR_IDLE_BASE_ID",
    IdleVariations = {
        "rbxassetid://YOUR_LOOK_AROUND_ID",
        "rbxassetid://YOUR_STRETCH_ID",
        -- ThÃªm cÃ¡c variations khÃ¡c náº¿u cÃ³
    },
    
    -- Timing
    MinIdleTime = 5,   -- Tá»‘i thiá»ƒu 5 giÃ¢y trÆ°á»›c khi variation
    MaxIdleTime = 10,  -- Tá»‘i Ä‘a 10 giÃ¢y
    
    -- Thresholds
    MovementThreshold = 0.1,  -- Speed Ä‘á»ƒ xÃ¡c Ä‘á»‹nh Ä‘á»©ng yÃªn
}

-- References
local character = script.Parent
local humanoid = character:WaitForChild("Humanoid")
local animator = humanoid:WaitForChild("Animator")

-- Variables
local idleBaseTrack = nil
local variationTracks = {}
local isIdle = false
local idleTimer = 0
local nextVariationTime = 0
local currentVariation = nil

-- Function: Load táº¥t cáº£ animations
local function loadAnimations()
    -- Load idle base
    local idleBaseAnim = Instance.new("Animation")
    idleBaseAnim.AnimationId = CONFIG.IdleBase
    idleBaseAnim.Name = "IdleBase"
    
    idleBaseTrack = animator:LoadAnimation(idleBaseAnim)
    idleBaseTrack.Priority = Enum.AnimationPriority.Idle
    idleBaseTrack.Looped = true
    
    -- Load variations
    for i, animId in ipairs(CONFIG.IdleVariations) do
        local varAnim = Instance.new("Animation")
        varAnim.AnimationId = animId
        varAnim.Name = "IdleVariation" .. i
        
        local varTrack = animator:LoadAnimation(varAnim)
        varTrack.Priority = Enum.AnimationPriority.Idle
        varTrack.Looped = false
        
        -- Event khi variation káº¿t thÃºc
        varTrack.Stopped:Connect(function()
            currentVariation = nil
            scheduleNextVariation()
        end)
        
        table.insert(variationTracks, varTrack)
    end
    
    print("Loaded", #variationTracks, "idle variations!")
end

-- Function: Random thá»i gian cho variation tiáº¿p theo
function scheduleNextVariation()
    nextVariationTime = math.random(CONFIG.MinIdleTime, CONFIG.MaxIdleTime)
    idleTimer = 0
    print("Next variation in", nextVariationTime, "seconds")
end

-- Function: Play random variation
local function playRandomVariation()
    if #variationTracks == 0 then return end
    if currentVariation then return end  -- Äang cháº¡y variation khÃ¡c
    
    -- Random chá»n 1 variation
    local randomIndex = math.random(1, #variationTracks)
    local varTrack = variationTracks[randomIndex]
    
    -- Play variation
    currentVariation = varTrack
    varTrack:Play(0.2)  -- Fade in
    
    print("Playing variation:", varTrack.Name)
end

-- Function: Start idle system
local function startIdle()
    if isIdle then return end
    
    isIdle = true
    idleBaseTrack:Play(0.2)
    scheduleNextVariation()
    
    print("Idle started")
end

-- Function: Stop idle system
local function stopIdle()
    if not isIdle then return end
    
    isIdle = false
    idleTimer = 0
    
    -- Stop base idle
    idleBaseTrack:Stop(0.2)
    
    -- Stop current variation náº¿u Ä‘ang cháº¡y
    if currentVariation then
        currentVariation:Stop(0.1)
        currentVariation = nil
    end
    
    print("Idle stopped")
end

-- Load animations
loadAnimations()

-- Detect movement
humanoid.Running:Connect(function(speed)
    if speed > CONFIG.MovementThreshold then
        -- Äang di chuyá»ƒn
        stopIdle()
    else
        -- Äá»©ng yÃªn
        startIdle()
    end
end)

-- Detect jumping
humanoid.StateChanged:Connect(function(oldState, newState)
    if newState == Enum.HumanoidStateType.Jumping or
       newState == Enum.HumanoidStateType.Freefall then
        stopIdle()
    end
end)

-- Main loop: Update timer vÃ  trigger variations
game:GetService("RunService").Heartbeat:Connect(function(deltaTime)
    if not isIdle then return end
    if currentVariation then return end  -- Äang cháº¡y variation
    
    -- TÄƒng timer
    idleTimer = idleTimer + deltaTime
    
    -- Kiá»ƒm tra náº¿u Ä‘áº¿n lÃºc play variation
    if idleTimer >= nextVariationTime then
        playRandomVariation()
    end
end)

-- Báº¯t Ä‘áº§u vá»›i idle (character má»›i spawn thÆ°á»ng Ä‘á»©ng yÃªn)
task.wait(0.5)
startIdle()

print("================================")
print("   IDLE VARIATION SYSTEM")
print("   Stand still to see variations!")
print("================================")
```
