## Code Ä‘áº§y Ä‘á»§

```lua
--[[
    EmoteSystem5.lua
    Vá»‹ trÃ­: StarterPlayer > StarterPlayerScripts
    Chá»©c nÄƒng: 5 emote animations - E, 1, 2, 3, 4
--]]

local Players = game:GetService("Players")
local UserInputService = game:GetService("UserInputService")

local player = Players.LocalPlayer

-- âš ï¸ THAY Táº¤T Cáº¢ ANIMATION IDs Báº°NG ID Cá»¦A Báº N SAU KHI PUBLISH
local ANIMATION_IDS = {
    Wave = "rbxassetid://YOUR_WAVE_ID",
    Dance = "rbxassetid://YOUR_DANCE_ID",
    Point = "rbxassetid://YOUR_POINT_ID",
    Bow = "rbxassetid://YOUR_BOW_ID",
    Cheer = "rbxassetid://YOUR_CHEER_ID",
}

local KEYBINDS = {
    [Enum.KeyCode.E] = "Wave",
    [Enum.KeyCode.One] = "Dance",
    [Enum.KeyCode.Two] = "Point",
    [Enum.KeyCode.Three] = "Bow",
    [Enum.KeyCode.Four] = "Cheer",
}

local animationTracks = {}
local currentTrack = nil
local isPlayingEmote = false

local function setupAnimations(character)
    local humanoid = character:WaitForChild("Humanoid")
    local animator = humanoid:WaitForChild("Animator")

    animationTracks = {}

    for name, animId in pairs(ANIMATION_IDS) do
        local anim = Instance.new("Animation")
        anim.AnimationId = animId
        anim.Name = name

        local track = animator:LoadAnimation(anim)
        track.Priority = Enum.AnimationPriority.Action
        track.Looped = (name == "Dance")  -- Chá»‰ Dance loop

        track.Stopped:Connect(function()
            if currentTrack == track then
                currentTrack = nil
                isPlayingEmote = false
            end
        end)

        animationTracks[name] = track
    end

    print("Loaded 5 emote animations!")
end

local function playEmote(emoteName)
    if isPlayingEmote then
        -- Dá»«ng emote hiá»‡n táº¡i (trá»« Dance - cáº§n nháº¥n Q Ä‘á»ƒ dá»«ng)
        if currentTrack and currentTrack.Name ~= "Dance" then
            return  -- Äá»£i emote one-shot káº¿t thÃºc
        end
        if currentTrack and emoteName == "Dance" then
            return  -- Äang dance rá»“i
        end
    end

    local track = animationTracks[emoteName]
    if not track then return end

    -- Dá»«ng track hiá»‡n táº¡i
    if currentTrack and currentTrack.IsPlaying then
        currentTrack:Stop(0.1)
    end

    currentTrack = track
    isPlayingEmote = true
    track:Play(0.1)
    print("Playing:", emoteName)
end

local function stopCurrentEmote()
    if currentTrack and currentTrack.IsPlaying then
        currentTrack:Stop(0.2)
        currentTrack = nil
        isPlayingEmote = false
        print("Stopped emote")
    end
end

local function onCharacterAdded(character)
    setupAnimations(character)
end

player.CharacterAdded:Connect(onCharacterAdded)
if player.Character then
    onCharacterAdded(player.Character)
end

UserInputService.InputBegan:Connect(function(input, gameProcessed)
    if gameProcessed then return end

    local emoteName = KEYBINDS[input.KeyCode]
    if emoteName then
        playEmote(emoteName)
    end

    -- PhÃ­m Q Ä‘á»ƒ dá»«ng (Ä‘áº·c biá»‡t cho Dance)
    if input.KeyCode == Enum.KeyCode.Q then
        stopCurrentEmote()
    end
end)

print("=== 5 EMOTE SYSTEM ===")
print("E = Wave | 1 = Dance | 2 = Point | 3 = Bow | 4 = Cheer | Q = Stop")
print("======================")
```
